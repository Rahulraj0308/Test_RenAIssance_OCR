```python
import keras_ocr
import cv2
import matplotlib.pyplot as plt
import numpy as np

img_path = r"C:\Users\91988\Pictures\Screenshots\Screenshot 2025-03-22 110334.png"
img = cv2.imread(img_path)

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

pipeline = keras_ocr.pipeline.Pipeline()

predictions = pipeline.recognize([img])

resulted_text=[]
for text, box in predictions[0]:
    resulted_text.append(text)
    box = [(int(x), int(y)) for x, y in box]  
    cv2.polylines(img_rgb, [np.array(box)], isClosed=True, color=(0, 255, 0), thickness=2) 
    cv2.putText(img_rgb, text, (box[0][0], box[0][1] - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 0, 0), 2, cv2.LINE_AA)

print(resulted_text)

plt.figure(figsize=(10, 6))
plt.imshow(img_rgb)
plt.axis("off")
plt.show()
```
